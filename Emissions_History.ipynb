```python
# Import dependencies

import pandas as pd
import numpy as np
from config import db_password
from sqlalchemy import create_engine
```

```python
# Read csv

df=pd.read_csv('historic_emissions.csv')
df.info()
```

# Data Cleaning

```python
df
```

```python
# Drop the Data Source column
df=df.drop(columns=['Data source'])
df
```

```python
# Change years to rows

df2=pd.melt(df, id_vars=["Country", "Sector", "Gas", "Unit"], var_name="Year", value_name="Emission")
df2
```

```python
df2.shape
```

```python
#Check for missing values
df2.isnull().sum()
```

```python
# Drop all the rows with NaN values
clean_df=df2.dropna()
clean_df.shape
```

```python
# Double check for missing values
clean_df.isnull().sum()
```

```python
# Open a file in write mode
with open('Emissions_History.csv', 'w', encoding='utf-8') as f:
    
    # Save the dataframe to the file 
    clean_df.to_csv(f, index= False)
```

```python
#create the connection to the PostgreSQL database
db_string = f"postgresql://postgres:{db_password}@127.0.0.1:5432/Emission_History"

#create the db engine
engine = create_engine(db_string)
    
# Load our datafrme into sql
clean_df.to_sql(name='Emissions_History', con=engine, index= False, if_exists='replace')
```
