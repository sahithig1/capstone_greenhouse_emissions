```python
# Import dependencies

import pandas as pd
import numpy as np
from config import db_password
from sqlalchemy import create_engine
```

```python
# Read csv

df=pd.read_csv('USA_HistoricEmissions.csv')
df.info()
```

## Data Cleaning

```python
df
```

```python
# Drop the Data Source column
df=df.drop(columns=['Data source', 'Country'])
df
```

```python
# Change years from columns to rows

usa_em_df=pd.melt(df, id_vars=["Sector", "Gas", "Unit"], var_name="Year", value_name="Emission")
usa_em_df
```

```python
usa_em_df.shape
```

```python
#Check for missing values
usa_em_df.isnull().sum()
```

```python
# Open a file in write mode
with open('USA_Emissions_History.csv', 'w', encoding='utf-8') as f:
    
    # Save the dataframe to the file 
    usa_em_df.to_csv(f, index= False)
```

```python
#create the connection to the PostgreSQL database
db_string = f"postgresql://postgres:{db_password}@127.0.0.1:5432/GHG_emissions"

#create the db engine
engine = create_engine(db_string)
    
# Load our datafrme into sql
usa_em_df.to_sql(name='USA_Emissions_History', con=engine, index= False, if_exists='replace')
```
