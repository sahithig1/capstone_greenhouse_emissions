```python
# Import dependencies

from geopy.geocoders import Nominatim
import pandas as pd
```

```python
# Read csv

df= pd.read_csv('CH4 Emissions.csv')
```

```python
# Add latitude and longtitude

def get_lat_long(country):
    geolocator= Nominatim(user_agent= 'my_app')
    location = geolocator.geocode(country)
    return(location.latitude, location.longitude)
```

```python
df['Latitude'] = df['Country'].apply(lambda x: get_lat_long(x)[0])
df['Longitude'] = df['Country'].apply(lambda x: get_lat_long(x)[1])
```

```python
# Drop columns

CH4_df=df.drop(columns=['Data source', 'Sector'])
CH4_df.tail()
```

```python
# Open a file in write mode
with open('CH4_Emissions.csv', 'w', encoding='utf-8') as f:

    # Save the dataframe to the file 
    CH4_df.to_csv(f, index= False)
```
