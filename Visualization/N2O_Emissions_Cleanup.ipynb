```python
# Import dependencies

from geopy.geocoders import Nominatim
import pandas as pd
```

```python
# Read csv

df=pd.read_csv('N2O Emissions.csv')
```

```python
df.info()
```

```python
# Add latitude and longtitude

def get_lat_long(country):
    geolocator= Nominatim(user_agent= 'my_app')
    location = geolocator.geocode(country)
    return(location.latitude, location.longitude)
```

```python
df['Latitude'] = df['Country'].apply(lambda x: get_lat_long(x)[0])
df['Longitude'] = df['Country'].apply(lambda x: get_lat_long(x)[1])
```

```python
# Drop columns

N2O_df=df.drop(columns=['Data source', 'Sector'])
N2O_df.head()
```

```python
# Open a file in write mode

with open('N2O_Emissions.csv', 'w') as f:
    
    # Same the dataframe to the file
    N2O_df.to_csv(f, index= False)
```
