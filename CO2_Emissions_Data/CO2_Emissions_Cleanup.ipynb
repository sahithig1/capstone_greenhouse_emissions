```python
# Import dependencies

import pandas as pd
import numpy as np
from config import db_password
from sqlalchemy import create_engine
```

```python
# Read csv

df=pd.read_csv('CO2_historical_emissions.csv')
```

```python
# Check for null values per column

df.isnull().sum()
```

```python
df=df.dropna()
df.size
```

```python
df.shape
```

```python
# Dropping the 'World' and "European Union (27)" rows.
# All countried that are a part of the European Union are already present in the dataframe

CO2_df = df.drop(df[(df['Country']=='World')|(df['Country']=='European Union (27)')].index)
CO2_df
```

```python
CO2_df.size
```

```python
CO2_df.shape
```

```python
CO2_df
```

```python
# Drop columns

CO2_df=CO2_df.drop(columns=['Data source'])
CO2_df.head()
```

```python
# Change years from columns to rows

CO2_em_df=pd.melt(CO2_df, id_vars=["Country", "Sector", "Gas", "Unit"], var_name="Year", value_name="Emission")
CO2_em_df
```

```python
# Open a file in write mode
with open('CO2_Emission.csv', 'w', encoding='utf-8') as f:
    
    # Save the dataframe to the file
    CO2_em_df.to_csv(f, index= False)
```

```python
#create the connection to the PostgreSQL database
db_string = f"postgresql://postgres:{db_password}@127.0.0.1:5432/GHG_emissions"

#create the db engine
engine = create_engine(db_string)
    
# Load our datafrme into sql
CO2_em_df.to_sql(name='CO2_Emissions_History', con=engine, index= False, if_exists='replace')
```
